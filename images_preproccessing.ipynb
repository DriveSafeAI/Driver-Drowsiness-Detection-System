# **Images Preprocessing**<br>

This code loads images from each class, converts them to grayscale, and saves them into a new organized directory structure.


<img src="https://cdn-icons-png.flaticon.com/512/9423/9423110.png" width="200">

```python
import os
import cv2
from glob import glob

base_dir = r"C:\Users\ToP NeT\Desktop\drowsiness_detection_dataset"

output_dir = r"C:\Users\ToP NeT\Desktop\dataset_preprocessed"
os.makedirs(output_dir, exist_ok=True)

for split in ["train", "test"]:
    split_dir = os.path.join(base_dir, split)
    output_split_dir = os.path.join(output_dir, split)
    os.makedirs(output_split_dir, exist_ok=True)

    for class_name in ["awake", "drowsy", "absent"]:
        class_dir = os.path.join(split_dir, class_name)
        output_class_dir = os.path.join(output_split_dir, class_name)
        os.makedirs(output_class_dir, exist_ok=True)

        image_files = glob(os.path.join(class_dir, "*.*"))

        for img_path in image_files:
                img = cv2.imread(img_path)
                if img is None:
                    print(f" error: {img_path}")
                    continue

                # Convert image to grayscale
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

                # Save the processed image to the new directory
                img_name = os.path.basename(img_path)
                save_path = os.path.join(output_class_dir, img_name)
                cv2.imwrite(save_path, gray)
```
