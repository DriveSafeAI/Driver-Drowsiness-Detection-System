**3- Extracting the dataset:**
Since the dataset is stored as a ZIP file, we first extract it into Colab. This step makes the training and testing folders accessible for model training. After extraction, we can browse the dataset structure and ensure that the images are well organized.


````python
from zipfile import ZipFile

zip_path = "/content/drive/MyDrive/drowsiness_detection_system_dataset.zip"
extract_path = "/content/dataset"

with ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
````


**4- Installing and importing YOLO library:**
To train our classification model, we need the Ultralytics YOLO library. Installing it in Colab ensures that all the required dependencies are available. After installation, we can import the library and prepare it for training.


````python
#Ultralytics YOLO
!pip install ultralytics

# import YOLO
from ultralytics import YOLO
````

**5- Training the model on Colab Pro with A100 GPU:**
In this stage, the model is trained on Google Colab Pro using a high-performance A100 GPU. The training process involves feeding the model with labeled images from the dataset so it can learn to distinguish between the three classes: **Awake**, **Drowsy**, and **Absent**. Over the course of 100 epochs, the model adjusts its internal parameters to minimize the loss function, gradually improving its ability to correctly classify each image. Continuous monitoring of training and validation metrics ensures that the model is learning effectively and not overfitting, resulting in a robust model capable of generalizing to new, unseen data.


````python
model = YOLO("yolo11n-cls.pt")

model.train(
    data="/content/dataset/drowsiness_detection_system_dataset",
    epochs=100,
    batch=32,
    imgsz=640,
    device=0,
    project="/content/drive/MyDrive/yolo_training",
    name="exp"
)
````
